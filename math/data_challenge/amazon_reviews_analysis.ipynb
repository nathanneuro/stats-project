```python
import pandas as pd
import gzip
import timeit
import numpy as np



def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)


def Load_Chunk(chunk, path, chunksize):
  start_time = timeit.default_timer()
  dfDict = {}
  # starting at chunksize*(chunk-1) going to chunksize*chunk  
  max = chunksize * chunk
  i = chunksize * (chunk - 1)
  if i < max:
    for d in parse(path):
      dfDict[i] = d
      i += 1
      if i>max:
        break

  df = pd.DataFrame.from_dict(dfDict, orient='index')
  
  df['reviewTime'] = pd.to_datetime(df['reviewTime']) # Convert to datetime or SQL fails
  # Convert 'helpful' list to str
  # df['helpful'] = df['helpful'].apply(str)
  
  print("Loaded Chunk #", chunk, "Total time elapsed: ", int(timeit.default_timer() - start_time), " seconds")
  return df








destination = '/Volumes/Storage/Lab data/Data Analysis/samplereviews.db'
global chunk
chunk = 1
df = Load_Chunk(chunk, '/Volumes/Storage/Lab data/Data Analysis/reviews_electronics.json.gz', 10000)
print(df.head())
print("Chunk #", chunk)
  

```
